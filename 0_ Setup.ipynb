# Download all files from Zenodo (~10 minutes)

```python
import requests, os, zipfile

# Fetch record metadata
url = f"https://zenodo.org/api/records/14502774"
files = requests.get(url).json()['files']

# Download files
for file in files:
    print(f"Downloading {file['key']}...")
    r = requests.get(file['links']['self'], stream=True)
    with open(os.path.join(root, file['key']), "wb") as f:
        for chunk in r.iter_content(8192):
            f.write(chunk)
print("All files downloaded!")

# Uncompress the ZIP file
with zipfile.ZipFile(root+'/data.zip', 'r') as zip_ref:
    zip_ref.extractall('data')

# Delete the ZIP file
os.remove(root+'/data.zip')
```
